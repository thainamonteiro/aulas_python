## Como percorrer um dataset usando loops (for)

Para isso, vocÃª pode usar a ideia de um loop, onde ele vai percorrer cada linha do seu dataset. Esse conceito vai ser bastante utilizando, principalmente na plotagem de um grÃ¡fico. No exemplo abaixo estamos percorrendo todo dataset (linha por linha), porÃ©m, "armazenando" (usando uma lista) apenas a coluna "OFFENSE_CODE"

```python
lista = []
for offense_code in df["OFFENSE_CODE"]:
    lista.append(offense_code)
```

## Criando um contador sem usar funÃ§Ãµes do prÃ³prio pandas

Esse cÃ³digo Ã© um exemplo do que vocÃª pode fazer sem usar o pandas. Ã‰ uma funÃ§Ã£o similar ao value_count, porÃ©m, estou apenas contando um elemento do meu dataset. No caso, quero contar as vezes que o DISTRICT = D14 aparece no meu dataset. VocÃª pode usar o pandas, mas isso Ã© apenas para ter uma noÃ§Ã£o de como fazer de outra forma

```python
count_d4 = 0 #Iniciando um contador com 0 (declarando a variavel)
for district in df["DISTRICT"]: #Iniciando o for para percorrer toda a coluna DISTRICT do nosso dataset
    
    # Esse trecho diz que a cada district D4, vamos incrementar o nosso contador
    if district == "D4":
        count_d4 += 1
        
print('NÃºmero de elementos D4: ', count_d4)
```

Mesmo principio do contador mostrado, porÃ©m, quero contar todas as ocorrencias dos elementos (Esse Ã© a mesma coisa do value count do pandas). A ideia aqui Ã© usar uma lista chave: valor. Esse tipo de estrutura funciona da seguinte forma (como um dicionÃ¡rio)

- {'dia': 'significado da palavra dia',
 'noite': 'significado da palavra noite'}
 
Fazendo uma analogia a um dicionÃ¡rio, eu posso saber o significado da palavra dia apenas procurando essa palavra no dicionÃ¡rio. Em Python, vocÃª acessaria dessa forma:

- lista['dia'], onde 'dia' Ã© uma chave para procurar um determinado valor.

```python
count = {}
for district in df["DISTRICT"]:
    if district not in count:
        count[district] = 0
    count[district] += 1
```

## Criando grÃ¡ficos de Pizza
Para fazer isso com matplotlib Ã© bem simples:

- plt.pie("Aqui vÃ£o os seus dados", labels="Nome dos seus dados", colors="Cor de cada um no grafico")


```python

plt.pie([10, 2, 3], labels = ['malware', 'trusted', 'riskware'], 
            colors = ['red', 'blue', 'green'], 
            autopct='%1.1f%%', startangle = 90)

plt.title("SeguranÃ§a digital")
plt.show()
```

Com isso, vocÃª pode mostrar o mesmo grÃ¡fico para ver o nÃºmero de ocorrÃªncias que vocÃª tem:

```python
count = {}
for district in df["DISTRICT"]:
    if district not in count:
        count[district] = 0
    count[district] += 1
```

```python
#count.values() <- Esse mostra apenas os valores de dentro do dicionario chave: valor
#count.items() <- Esse mostra apenas as categorias

# Note que aqui nÃ£o definimos COLORS. Isso quer dizer que isso serÃ¡ definido automÃ¡tico
plt.pie(count.values(), labels = count.items(), 
            autopct='%1.1f%%', startangle = 90)

plt.title('Crimes (DISTRICT)')
plt.show()
```

Talvez alguns grÃ¡ficos nÃ£o demonstrem bem os dados. Por isso, vocÃª precisa tentar modificar o tipo para ver qual representa melhor a relaÃ§Ã£o do seu dataset. Vou mostrar os mesmos dados acima, porÃ©m, com um grÃ¡fico diferente:

```python
# Aqui eu faÃ§o a contagem com o pandas
numbers_of_district = df["DISTRICT"].value_counts()
# Aqui eu chamo o meu grafico de barras
numbers_of_district.plot(kind='bar')

# Titulo do grafico
plt.title(" NÃºmeros de crimes por distrito (Boston)")

# Texto eixo x
plt.xlabel(" Distritos")

# Texto eixo y
plt.ylabel(" NÃºmero de ocorrÃªncias (crimes)")
# Aqui eu mostro o meu grafico de barras
plt.show()
```

Pensando no seu dataset, como vocÃª pode compara-lo? Tente fazer uma correlaÃ§Ã£o (em um grÃ¡fico) de contagem, ou seja, quantas ocorrÃªncias "mobile" e quantas mudanÃ§as "mobile" vocÃª teve em um determinado dia? 

Exemplo de um grÃ¡fico de barras:

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np



labels = ['02/02/2020', '03/02/2020', '04/02/2020']
numero_ocorrencias_mobile = [20, 50, 1]
numero_mudancas_mobile = [25, 4, 2]

x = np.arange(len(labels))  # PosiÃ§Ã£o dos textos
width = 0.35  # Tamanho da barra

fig, ax = plt.subplots() # Chamando os subgrÃ¡ficos

rects1 = ax.bar(x - width/2, numero_ocorrencias_mobile, width, label='ocorrencias') # Primeira barra
rects2 = ax.bar(x + width/2, numero_mudancas_mobile , width, label='mudanÃ§as') # Segunda barra


ax.set_ylabel(' OcorrÃªncias ')
ax.set_title(' Grupo de ocorrÃªncias')
ax.set_xticks(x)
ax.set_xticklabels(labels, size = 10)
ax.legend()
```
